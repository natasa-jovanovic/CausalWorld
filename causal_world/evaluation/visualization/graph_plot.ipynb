```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
data = pd.read_csv(r"C:\Users\psiml8\CausalWorld\scripts\experiments-pushing\0\7\model\0.monitor.csv")
frac_succ = data['fractional_success']
time = data['t']
```

```python
import numpy as np
averaged_data = []
for i in range(len(frac_succ)):
    if i < 100:
        averaged_data.append(0)
    else:
        averaged_data.append(np.mean(frac_succ[i - 100 : i]))
```

```python
plt.plot(time, averaged_data)
plt.show()
```

```python
plt.plot(time, frac_succ)
plt.show()
```
